```python
import numpy as np
ndim = 4
nsamp = 10000
np.random.seed(10)
A = np.random.rand(ndim,ndim)
cov = np.dot(A, A.T)
samps = np.random.multivariate_normal([0]*ndim, cov, size=nsamp)
A = np.random.rand(ndim,ndim)
cov = np.dot(A, A.T)
samps2 = np.random.multivariate_normal([0]*ndim, cov, size=nsamp)
```

```python
names = ["x%s"%i for i in range(ndim)]
labels =  ["x_%s"%i for i in range(ndim)]
samples = MCSamples(samples=samps,names = names, labels = labels)
samples2 = MCSamples(samples=samps2,names = names, labels = labels)
```

```python
#Triangle plot
g = plots.getSinglePlotter(width_inch=4, ratio=1)
g.plot_2d([samples, samples2], 'x1', 'x2')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
speed_of_light = 299792.458 ### micron GHz
labs = ("1 comp", "2 comp", r"1 $\beta=2$", r"2 $\beta=2$", "thick")

nptd = 10
npt = 10*nptd
xd = np.linspace(30,857,nptd)
sigd = 150
yd=xd+np.random.randn(nptd)*sigd
xd = speed_of_light/xd
yd = np.flipud(yd)


for lab in labs:
    

    plt.loglog(xd,yd, 'o')
    plt.errorbar(xd,yd,sigd,fmt='o')

    xlab = "wavelength [$\mu$m]"
    plt.xlabel(xlab, size='x-large')
    plt.ylabel("flux density [Jy]", size='x-large')

    plt.tick_params(labelsize='large')


    x  = np.linspace(30,857,npt)
    y=x+np.random.randn(npt)*30
    x = speed_of_light/x
    y = np.flipud(y)
    plt.plot(x,y,label=lab)
    plt.title('object')
    
plt.legend(loc='best')
```
